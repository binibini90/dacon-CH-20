```python
import pandas as pd
```

```python
spots = pd.read_csv('../data/tourist_spots.csv')
rests = pd.read_csv('../data/seoul_restaurant.csv')
```

```python
display(spots.info())
spots_c = spots[['관광지명_한글', '지번주소', '위도', '경도']].set_axis(['관광지명', '주소', '위도', '경도'], axis=1)
display(spots_c.head())
```

```python
display(rests.info())
rests_c = rests[['rest', 'post', 'latitude', 'longitude', 'category', 'score', 'count' ]].set_axis(['식음업장명', '주소', '위도', '경도', '분류', '평점', '리뷰수'], axis=1)
display(rests_c.head())
```

```python
# spots_c['주소']와 rests_c['주소'] 중 '구' 부분만 따로 컬럼('구')으로 추가
spots_c['구'] = spots_c['주소'].str.extract(r'서울\s+(\S+구)')
rests_c['구'] = rests_c['주소'].str.extract(r'서울\s+(\S+구)')

display(spots_c.head())
display(rests_c.head())
```
